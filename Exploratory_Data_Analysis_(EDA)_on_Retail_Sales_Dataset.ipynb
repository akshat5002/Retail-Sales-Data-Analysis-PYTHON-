```python
import pandas as pd
```

```python
df=pd.read_csv('/content/sales_data.csv')
df.head(10)
```

BASIC EDA(exploratory data analysis)  steps:

```python
# Total sales
total_sales = df["Sales"].sum()
print("Total Sales:", total_sales)

# Total profit
total_profit = df["Profit"].sum()
print("Total Profit:", total_profit)

# Sales by category
category_sales = df.groupby("Category")["Sales"].sum()
print("Sales by Category:")
print(category_sales)

# Best selling product
top_product = df.groupby("Product")["Sales"].sum().idxmax()
print("Top Product:", top_product)

# Region wise sales
region_sales = df.groupby("Region")["Sales"].sum()
print("Region Sales:")
print(region_sales)
```

```python
import matplotlib.pyplot as plt
# -----------------------------
# 1. Category Wise Sales
# -----------------------------
category_sales = df.groupby("Category")["Sales"].sum()
plt.figure()
category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
# -----------------------------


# 2. Region Wise Sales
# -----------------------------
region_sales = df.groupby("Region")["Sales"].sum()
plt.figure()
region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales Distribution by Region")
plt.ylabel("")
plt.show()

# -----------------------------
# 3. Top Product
# -----------------------------
product_sales = df.groupby("Product")["Sales"].sum().sort_values(ascending=False)
plt.figure()
product_sales.head(5).plot(kind='bar')
plt.title("Top 5 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

In this project, I analyzed retail sales data using Python to derive meaningful business insights.
The analysis revealed that the Electronics category is the top revenue generator, indicating strong customer demand in this segment.
Region-wise analysis showed that the West region contributes the highest sales, suggesting it as a key market for business growth.
Additionally, identifying top-selling products helps in better inventory and marketing decisions.
Overall, this analysis can support data-driven decision-making to improve sales performance and profitability.
